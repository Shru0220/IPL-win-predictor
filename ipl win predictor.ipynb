```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')
```

```python
total_score_df = deliveries.groupby(['ID','innings']).sum()['total_run'].reset_index()
```

```python
total_score_df
```

```python
total_score_df = total_score_df[total_score_df['innings'] == 1]
```

```python
total_score_df
```

```python
match_df = match.merge(total_score_df[['ID','total_run']], left_on='ID', right_on='ID')
```

```python
match_df
```

```python
match_df['Team1'].unique()
```

```python
teams = [
'Chennai Super Kings',
'Delhi Capitals',
'Gujarat Titans',
'Kolkata Knight Riders',
'Lucknow Super Giants',
'Mumbai Indians',
'Kings XI Punjab',
'Rajasthan Royals',
'Royal Challengers Bangalore',
'Sunrisers Hyderabad'
]
```

```python
match_df['Team1'].value_counts()
```

```python
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

```

```python
match_df = match_df[match_df['Team1'].isin(teams)]
match_df = match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
#deliveries_df =deliveries_df['BattingTeam'].str.replace('Delhi Daredevils','Delhi Capitals')
#deliveries_df =deliveries_df['BattingTeam'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df.shape
```

```python
def bowling_team(row):
    if row['WinningTeam'] == row['Team1']:
        return row['Team2']
    else:
        return row['Team1']
```

```python
match_df['bowling_team'] = match_df.apply(bowling_team, axis=1)
```

```python
match_df.head()
```

```python
match_df = match_df[['ID','City','WinningTeam','bowling_team','total_run']]
```

```python
match_df
```

```python
deliveries_df = match_df.merge(deliveries,on='ID')
```

```python
deliveries_df
```

```python
deliveries_df =  deliveries_df[deliveries_df['innings'] == 2]
```

```python
deliveries_df.shape
```

```python
deliveries_df['current_score'] = deliveries_df.groupby('ID')['total_run_y'].cumsum()
```

```python
deliveries_df
```

```python
deliveries_df['runs_left'] =deliveries_df['total_run_x'] - deliveries_df['current_score']
```

```python
deliveries_df
```

```python
deliveries_df['balls_left'] = 126 - (deliveries_df['overs']*6 + deliveries_df['ballnumber'])
```

```python
deliveries_df
```

```python
deliveries_df['player_out'] = deliveries_df['player_out'].fillna("0")
```

```python
deliveries_df
```

```python
deliveries_df['player_out'] = deliveries_df['player_out'].apply(lambda x:x if x == "0" else "1")
```

```python
deliveries_df
```

```python
deliveries_df['player_out'] = deliveries_df['player_out'].astype('int')
```

```python
deliveries_df
```

```python
wickets = deliveries_df.groupby('ID').cumsum(numeric_only=True)['player_out'].values
```

```python
deliveries_df['wickets'] = 10 - wickets
```

```python
deliveries_df
```

```python
deliveries_df.tail()
```

```python
deliveries_df['crr'] = (deliveries_df['current_score']*6)/(120 - deliveries_df['balls_left'])
```

```python
deliveries_df
```

```python
deliveries_df['rrr'] = (deliveries_df['runs_left']*6)/deliveries_df['balls_left']
```

```python
deliveries_df
```

```python
def result(row):
    return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
deliveries_df['result'] = deliveries_df.apply(result,axis=1)
```

```python
deliveries_df.head()
```

```python
final_df = deliveries_df[['BattingTeam','bowling_team','City','runs_left','balls_left','wickets','total_run_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.info()
```

```python
final_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.info()
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','bowling_team','City'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
